```python
## web-based loader
## whle laoding , cleaning the data with bs4 parameters
## for example, you will give a website, it should take or load the information shown on the webpage

from langchain_community.document_loaders import WebBaseLoader
import bs4

web_loader = WebBaseLoader(web_path=("https://geoffrey-geofe.medium.com/tokenization-vs-embedding-understanding-the-differences-and-their-importance-in-nlp-b62718b5964a"),
                           bs_kwargs=dict(parse_only = bs4.SoupStrainer(
                               class_=("post-title","post-content", "post-header"))
                           ))

```

```python
web_loader.load()
```
